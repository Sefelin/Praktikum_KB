### Import Package

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read Dataset

```python
df = pd.read_csv("netflix.csv")
df
```

## 1. Grafik Batang

```python
x = df.country[0:10]
y = df.duration[0:10]

plt.bar(x,y,color='Pink')
plt.title("Perbandingan durasi film setiap negara")
plt.xticks(rotation=90)
plt.xlabel("Negara film")
plt.ylabel("Durasi Film")
plt.show()
```

Grafik batang di atas, dapat disimpulkan bahwa Negara Brazil memiliki durasi paling lama diantara 7 negara, dengan durasi 4 Season dan yang memili durasi paling sedikit Negara Mexico dengan durasi 93 min.

## 2. Scatter Plot


```python
sns.scatterplot(data = df, x='type', y='release_year')
plt.title("Hubungan type film dengan tahun rilis")
plt.xticks(rotation=90)
plt.show()
```

Pada Grafik diatas menunjukkan tipe film "Movie" lebih banyak dirilis pada tahun 1920 keatas, sedangkan tipe film "TV show" mulai banyak dirilis pada tahun 1960 keatas.

## 3. Grafik Pie

```python
data_pie = df['rating'].value_counts().rename_axis('rating').reset_index(name='Jumlah')
data_pie
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title("Reting film ")
plt.pie(df['rating'].value_counts(), labels=data_pie['rating'], autopct='%.1f%%', startangle=0)
plt.show()
```

Grafik pie diatas menampilkan persebaran rating-rating film dimana rating TV-MA memiliki film terbanyak diantara 13 rating film lainnya. Dan yang meduduki penyebaran rating terndah yaitu NC-17.
