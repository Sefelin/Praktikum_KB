##  Data Cleaning

Menangani Nilai Null


```python
# Mengecek data null
df.isna().sum()
```

```python
imputer_string = SimpleImputer(strategy="most_frequent") #modus
```

```python
df['director'] = imputer_string.fit_transform(df[["director"]])
df['cast'] = imputer_string.fit_transform(df[["cast"]])
df['country'] = imputer_string.fit_transform(df[["country"]])
df['date_added'] = imputer_string.fit_transform(df[["date_added"]])
df['release_year'] = imputer_string.fit_transform(df[["release_year"]])
df['rating'] = imputer_string.fit_transform(df[["rating"]])

```

```python
imputer_float = SimpleImputer(strategy="mean")
```

```python
df['release_year'] = imputer_float.fit_transform(df[["release_year"]])
```

```python
df.isna().sum()
```
